## Importing the Data

## Converting Wrong Data Types - Price

```python
two_dot_percentage = ((products.price.str.contains("\d+\.\d+\.\d+").value_counts()[1] / products.shape[0])*100).round(2)
print(f"The 2 dot problem represents {two_dot_percentage}% of the rows in our DataFrame")
```

```python
numeric_price_products = products.loc[products.price.str.contains("\d+\.\d+\.\d+", na=False)]
two_dot_order_ids_list = numeric_price_products["type"]
products = products.loc[~products.type.isin(two_dot_order_ids_list)]
```

```python
products['price'] = pd.to_numeric(products['price'])
```
