# Merge Tables

```python
oolp = orderlines.merge(orders, how='left',left_on="id_order",right_on="order_id").merge(products, how="left",on="sku")
oolp
```

## Drop Columns

```python
columns_to_drop = ["product_id", "order_id", "created_date"]
oolp = oolp.drop(columns=columns_to_drop)
```

## Rename Columns

```python
columns_to_rename = ({"id":"orderlines_id", "price":"base_price"})
oolp = oolp.rename(columns=columns_to_rename)
```

# Calculate discounts

```python
oolp["discounts"] = oolp["base_price"] - oolp["unit_price"]
```

```python
oolp.sort_values(by = "discounts", ascending = False)
```

```python
oolp["discount_percentage"] = oolp["discounts"] / oolp["base_price"] * 100
```

```python
oolp['discount_percentage'].mean()
```

```python
oolp["date"] = pd.to_datetime(oolp["date"])
```

```python
oolp["year"] = oolp["date"].dt.year
oolp["month"] = oolp["date"].dt.month
```

```python
# calculate monthly revenue taking into consideration the sales price * amount of items sold
oolp.groupby(["year", "month"])["product_times_quantity"].sum().plot.bar()
```

```python
#monthly discount percentage - seems constant (minor fluctuation)
oolp.groupby(["year", "month"])["discount_percentage"].mean().plot.bar()
```

```python
oolp.groupby(["year", "month"])["discounts"].sum().plot.bar()
```

# Best Selling Products


```python
agg_dict = {
    'unit_price': 'sum',
    'orderlines_id': 'count',
    'discounts': 'sum',
    'product_quantity': 'sum'
}

agg_result = oolp.groupby('sku').agg(agg_dict)
```

```python
#products that overall generated less revenue than their discounts
agg_result.loc[agg_result['discounts'] > agg_result['unit_price']]["unit_price"].sum()
```

```python
agg_result
```

```python
#top 10 products that generated more discounts than revenue
top_10_loss = agg_result.loc[agg_result['discounts'] > agg_result['unit_price']].sort_values("unit_price",ascending = False).head(10)
```

```python
top_10_loss_df = pd.DataFrame(top_10_loss)
```

```python
top_10_loss_df.reset_index(inplace = True)
```

```python
top_10_loss_df = top_10_loss_df.drop("sku", axis=1)
```

```python
top_10_loss_df = top_10_loss_df.drop("orderlines_id",axis=1)
```

```python
columns_to_rename_2 = ({"unit_price":"revenue", "discounts":"cumulative_discount","product_quantity":'amount_sold'})
top_10_loss_df = top_10_loss_df.rename(columns=columns_to_rename_2)
```

```python
top_10_loss_df["loss"] = top_10_loss_df["revenue"] - top_10_loss_df["cumulative_discount"]
```

# Findings

1.   All year long appx. 20% discount on items - constant
2.   The sum of discounts and the sum of sales seem to move perfectly together which means:
3.   Sale spikes seem to be due to seasonality and not due to discounts
4.   In the case of 367 items the sum of all discounts are more than the income
      from them - these products account for 7%
5.   Revenue: 132955.85 of the revenue - accounts for less **than 1% of revenue**
6.   Average product value 161 for items in actual sales, 667 for product catalog
7.   Items in shopping basket - why are 50% of customers not converted
8.   Top 50 best selling items account for **24% of all orders, but only bring in 16.3% of the revenue**
9.  Top 100 best selling items account for **35.22% of all orders, but only bring in 26.61% of the revenue**
